```python
import matplotlib.pyplot as plt
```

```python

x = [5,10,15,20,25,30,35,40]
y = [
13.23733639,
12.56933801,
12.3535384,
12.26782066,
12.25892821,
12.23712957,
12.22455577,
12.22975413]
ci = [
0.116526500,
0.071128902,
0.040813574,
0.025072512,
0.024406759,
0.022860783,
0.020939605,
0.021477009
]
y_b = [a - b/2 for a, b in zip(y,ci)]
y_t = [a + b/2 for a, b in zip(y,ci)]
```

```python
fig, ax = plt.subplots()
ax.plot(x,y,'-o')
ax.fill_between(x, y_b, y_t, alpha=.1)
ax.grid()
ax.set_xlabel('Plowing Speed [mph]')
ax.set_ylabel('Mean Travel Time')
plt.show()
```
